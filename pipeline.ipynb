```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer , StandardScaler, OneHotEncoder , LabelEncoder ,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
def encoding_user_profile_tags(df):
    df['tags_list'] = df['User_profile_tags'].fillna("NaN").str.split(',')
    df.head()
    mlb = MultiLabelBinarizer()
    tags = mlb.fit_transform(df['tags_list'])
    profile_tags_df = pd.DataFrame(tags, columns=mlb.classes_, index=df.index)
    df_final = pd.concat([df, profile_tags_df], axis=1)
    df = df_final
    df.drop('tags_list', axis=1, inplace=True)
    df.drop('User_profile_tags',axis=1,inplace=True)
    return df
```

```python
def extracting_hours(df):
    df['Timestamp'] = pd.to_datetime(df['Timestamp'].astype(str).str[:14], format='%Y%m%d%H%M%S')
    df['Hours'] = df['Timestamp'].dt.hour 
    df.drop('Timestamp',axis=1,inplace=True)
    return df
```

```python
column_mapping = {
    10006: "interest_news",
    10024: "interest_education",
    10031: "interest_automobile",
    10048: "interest_real_estate",
    10052: "interest_IT",
    10057: "interest_electronic_game",
    10059: "interest_fashion",
    10063: "interest_entertainment",
    10067: "interest_luxury",
    10074: "interest_home_and_lifestyle",
    10075: "interest_health",
    10076: "interest_food",
    10077: "interest_divine",
    10079: "interest_motherhood_parenting",
    10083: "interest_sports",
    10093: "interest_travel_outdoors",
    10102: "interest_social",
    10684: "Inmarket_3c_product",
    11092: "Inmarket_appliances",
    11278: "Inmarket_clothing_shoes_bags",
    11379: "Inmarket_Beauty_Personal_Care",
    11423: "Inmarket_household_home_improvement",
    11512: "Inmarket_infant_mom_products",
    11576: "Inmarket_sports_item",
    11632: "Inmarket_outdoor",
    11680: "Inmarket_health_care_products",
    11724: "Inmarket_luxury",
    11944: "Inmarket_real_estate",
    13042: "Inmarket_automobile",
    13403: "Inmarket_finance",
    13496: "Inmarket_travel",
    13678: "Inmarket_education",
    13776: "Inmarket_service",
    13800: "interest_art_photography_design",
    13866: "interest_online_literature",
    13874: "Inmarket_electronic_game",
    14273: "interest_3c",
    16593: "Inmarket_book",
    16617: "Inmarket_medicine",
    16661: "Inmarket_food_drink",
    16706: "interest_culture",
    16751: "interest_sex",
    10110: "Demographic_gender_male"
}

def map_browser(agent):
    browsers = ['edge', 'trident', 'chrome', 'firefox', 'safari', 'opera']
    for browser in browsers:
        if browser in agent.lower():
            return browser
    return 'other'

def map_os(agent):
    os_list = ['windows', 'linux', 'mac os x']
    for os in os_list:
        if os in agent.lower():
            return os
    return 'other'
    
def apply_mappings(df):
    df['OS'] = df['User-Agent'].astype(str).apply(map_os)
    df['Browser'] = df['User-Agent'].astype(str).apply(map_browser)
    df.drop(['User-Agent'],axis=1,inplace=True)
    df.rename(columns={str(k): v for k, v in column_mapping.items()}, inplace=True)
    return df
```

```python
def ad_area(df):
    df['Adslotarea']=df['Adslotheight']*df['Adslotwidth']
    df.drop(['Adslotheight'],axis=1,inplace=True)
    df.drop(['Adslotwidth'],axis=1,inplace=True)
    return df
```
